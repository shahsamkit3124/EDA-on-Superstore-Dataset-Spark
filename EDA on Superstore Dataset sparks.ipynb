```python
# Perform groupby operation on 'Category' and calculate the sum of 'Profit' and 'Sales'
grouped_data = super_store.groupby('Category')[['Profit', 'Sales']].sum()

# Plot the aggregated data as a bar chart
grouped_data.plot(kind='bar')
plt.title('Total Profit and Sales by Category')
plt.xlabel('Category')
plt.ylabel('Amount')
plt.show()

```
